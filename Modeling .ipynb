the timestamp column on the 3 datasets are not stored as dates. Let's fix that

```python
def convert_to_datetime(data: pd.DataFrame=None, column: str=None):
    dummy=data.copy()
    dummy[column]=pd.to_datetime(dummy[column], format = '%Y-%m-%d %H:%M:%S')
    return dummy
```

### Merging data 

Client wants the model to predict on hourly basis. we have to change the timestamps on all the datasets to hourly 

```python
from datetime import datetime
def to_hourly(data: pd.DataFrame=None, column: str=None):
    dummy=data.copy()
    ts=dummy[column].tolist()
    ts=[i.strftime('%Y-%m-%d %H:00:00') for i in ts]
    ts=[datetime.strptime(i,'%Y-%m-%d %H:00:00') for i in ts]
    dummy[column]=ts
    return dummy 
```

```python
df_sales=to_hourly(df_sales, 'timestamp')
df_sales.head()
```

```python
df_stock=to_hourly(df_stock, 'timestamp')
df_stock.head()
```

```python
df_temp=to_hourly(df_temp, 'timestamp')
df_temp.head()
```

```python
#grouping sales data by 'timestamp','product_id'
sales_agg=df_sales.groupby(['timestamp','product_id']).agg({'quantity':'sum'}).reset_index()
sales_agg.head()
```

```python
#grouping stock data by 'timestamp','product_id'
stock_agg=df_stock.groupby(['timestamp','product_id']).agg({'estimated_stock_pct':'mean'}).reset_index()
stock_agg.head()
```

```python
#grouping temperature data by 'timestamp','product_id'
temp_agg=df_temp.groupby(['timestamp']).agg({'temperature':'mean'}).reset_index()
temp_agg.head()
```

```python
#merging data
merged_df=stock_agg.merge(sales_agg,on=['timestamp','product_id'],how='left')
merged_df.head()
```

```python
merged_df=merged_df.merge(temp_agg,on='timestamp',how='left')
merged_df.head()
```

### Feature Engineering 

```python
#exploding timestamp column into day of the week, day of the month and hour 
merged_df['timestamp_day_of_month']=merged_df['timestamp'].dt.day
merged_df['timestamp_day_of_week']=merged_df['timestamp'].dt.dayofweek
merged_df['timestamp_hour']=merged_df['timestamp'].dt.hour
merged_df.drop(columns=['timestamp'],inplace=True)
merged_df.head()
```

```python
# Changing category column to numerical values 
merged_df=pd.get_dummies(merged_df,columns=['category'])
merged_df.head()
```

```python
merged_df.info()
```

```python
#dropping product id column 
merged_df.drop(columns=['product_id'],inplace=True)
merged_df.head()
```

### Modeling

```python
# Creating features and target variables
X=merged_df.drop(columns=['estimated_stock_pct'])
y=merged_df['estimated_stock_pct']
print(X.shape)
print(y.shape)
```

```python
# Defining number of folds and split ratio 
K=10
split=0.80
```

```python
Accuracy = []

for fold in range(0, K):

  # Instantiate algorithm
  model = RandomForestRegressor()
  scaler = StandardScaler()

  # Create training and test samples
  X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=split, random_state=42)

  # Scale X data, we scale the data because it helps the algorithm to converge
  # and helps the algorithm to not be greedy with large values
  scaler.fit(X_train)
  X_train = scaler.transform(X_train)
  X_test = scaler.transform(X_test)

  # Train model
  trained_model = model.fit(X_train, y_train)

  # Generate predictions on test sample
  y_pred = trained_model.predict(X_test)

  # Compute accuracy, using mean absolute error
  mae = mean_absolute_error(y_true=y_test, y_pred=y_pred)
  Accuracy.append(mae)
  print(f"Fold {fold + 1}: MAE = {mae:.3f}")

print(f"Average MAE: {(sum(Accuracy) / len(Accuracy)):.2f}")
```

```python
features = [i.split("__")[0] for i in X.columns]
importances = model.feature_importances_
indices = np.argsort(importances)

fig, ax = plt.subplots(figsize=(10, 20))
plt.title('Feature Importances')
plt.barh(range(len(indices)), importances[indices], color='b', align='center')
plt.yticks(range(len(indices)), [features[i] for i in indices])
plt.xlabel('Relative Importance')
plt.show()
```
